```python
celsius = float(input("Enter your area temperature in f0...... :>"))
fahrenheit = (celsius * 9/5) + 32
print("Your temperature is in celsius :>",fahrenheit)
```

```python
fahrenheit = float(input("Enter the temperature in fahrenheit :>"))
celsius = (fahrenheit - 32)*5/9
print("Your temperature is in fahrenheit:>",round(celsius,2))
```

```python
celcius = float (input("Enter the temperature in celcius:>"))
kelvin = celcius + 273.15
print("This is your temperature in the kelvin :>",round (kelvin,3))
```
